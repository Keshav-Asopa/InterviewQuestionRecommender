```python
#importing libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
#loading data
data = pd.read_csv('assignment1.csv')
dataq = pd.read_csv('assignment2.csv')

X = data.iloc[:,0]
Y = dataq.iloc[:,:2]
Q = dataq.iloc[:,1]
X1 = data.iloc[:,1]
X2= data.iloc[:,2]
```

```python
# getting list of questions related to given topics
ques = []
c = 0
for i in X:
    for j in Y['Index']:
        if i==j:
            str = Y['question'][c]
            ques.append(str)
            c = c + 1


```

```python
#creating the matrix by vectorising the attributes
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(Q)
```

```python
#finding the cosine similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Build a 1-dimensional array with movie titles
titles = dataq['question']
indices = pd.Series(dataq.index, index=dataq['question'])
```

```python
# Function that get question recommendations based on the cosine similarity score of questions related to given topics
def genre_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
# here number inputs from (0-19) indicates the 19 topics given in the dataset in sequential manner 
print(genre_recommendations(10).head(10))
```
